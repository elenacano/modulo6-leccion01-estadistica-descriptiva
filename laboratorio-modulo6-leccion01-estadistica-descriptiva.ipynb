```python
def grafica_varianzas(df, columna):

    desviacion = df[columna].std()
    media = df[columna].mean()

    sns.histplot(x=columna, 
                    data=df,
                    bins=30, 
                    edgecolor="black", 
                    color="orange")

    # Añadir líneas de percentiles
    lista_cocientes = [1,-1,2,-2]
    for i in lista_cocientes:
        value = media + (desviacion*i)
        plt.axvline(value, color='green')
    plt.axvline(media, color='red')
    
    plt.title(f'Histograma con desviaciones para {columna}')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.legend()
```

- Varianza y desviación típica:

```python
import seaborn as sns
import matplotlib.pyplot as plt

def grafica_varianzas(df, columnas):
    # Crear una figura con subtramas, una para cada columna
    fig, axes = plt.subplots(len(columnas),1, figsize=(10, 10), sharey=True)

    # Iterar sobre las columnas y las subtramas correspondientes
    for i, columna in enumerate(columnas):
        desviacion = df[columna].std()
        media = df[columna].mean()

        # Graficar el histograma en la subtrama correspondiente
        sns.histplot(x=columna, 
                     data=df,
                     bins=30, 
                     edgecolor="black", 
                     color="orange",
                     ax=axes[i])

        # Añadir líneas de desviación estándar
        lista_cocientes = [1, -1, 2, -2]
        for j in lista_cocientes:
            value = media + (desviacion * j)
            axes[i].axvline(value, color='green')

        # Línea de la media
        axes[i].axvline(media, color='red')

        # Configurar título y etiquetas de cada subtrama
        axes[i].set_title(f'Histograma con desviaciones para {columna}')
        axes[i].set_xlabel('Valor')
        axes[i].set_ylabel('Frecuencia' if i == 0 else '')

    # Ajustar la presentación
    plt.tight_layout()
    plt.show()

# Llamar a la función para las columnas especificadas
grafica_varianzas(df, ["Horas Trabajadas", "Ventas Realizadas"])
```
